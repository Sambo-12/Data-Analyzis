## Means that exceed one time of standard deviation 

```python
def filter_group(group):
    
    #Spad1
    median_spad1 = group["Spad1"].mean()
    std_spad1 = group["Spad1"].std()
    lower1 = median_spad1 - std_spad1
    upper1 = median_spad1 + std_spad1
    #print(f"Group:\n{group}")
    print(f"Median Spad1: {median_spad1}")
    print(f"Lower: {lower1}")
    print(f"Upper: {upper1}") 
    
    #Spad2
    median_spad2 = group["Spad2"].mean()
    std_spad2 = group["Spad2"].std()
    lower2 = median_spad2 - std_spad2
    upper2 = median_spad2 + std_spad2

    #Spad3
    median_spad3 = group["Spad3"].mean()
    std_spad3 = group["Spad3"].std()
    lower3 = median_spad3 - std_spad3
    upper3 = median_spad3 + std_spad3

    filtered = group[
        (group["Spad1"] > lower1)  & (group["Spad1"] < upper1) &
        (group["Spad2"] > lower2)  & (group["Spad2"] < upper2) &
        (group["Spad3"] > lower3)  & (group["Spad3"] < upper3)
        ]

    return filtered

spad_filtered = spad.groupby("Nrates").apply(filter_group).reset_index(drop=True)
spad_filtered.groupby("Nrates")["Spad1"].describe()
```

```python
spad_filtered
```

```python
for column in ["Spad1", "Spad2", "Spad3"]: 
    corr, p_value = pearsonr(spad_filtered["Nrates"], spad_filtered[column])
    print("Pearson:", corr )
    print("p_value:", p_value)
```

```python
for column in ["Spad1", "Spad2", "Spad3"]:
    corr, p_value = pearsonr(spad_filtered["Nrates"], spad_filtered[column])
    print(f"Pearson correlation between Nrates and {column}: {corr:.8f}")
    print(f"P-value: {p_value:.8f}\n")
```

# Using Qualtile to clean data

```python
spad.shape
q1 = spad.groupby("Nrates")["Spad1"].quantile(0.25)
q3 = spad.groupby("Nrates")["Spad1"].quantile(0.75)
iqr = q3 - q1
lowerfence = q1- 1.5*iqr
upperfence = q1+ 1.5*iqr
print("lowerfence:", lowerfence)
print("upperfence:", upperfence)
```

```python
def outlier(group): 
    #define Q
    q1 = spad.groupby("Nrates")["Spad1"].quantile(0.25)
    q3 = spad.groupby("Nrates")["Spad1"].quantile(0.75)
    iqr1 = q3 - q1
    lowerfence = q1- 1.5*iqr1
    upperfence = q3+ 1.5*iqr1

    q2 = spad.groupby("Nrates")["Spad2"].quantile(0.25)
    q5 = spad.groupby("Nrates")["Spad2"].quantile(0.75)
    iqr2 = q5 - q2
    lowerfence2 = q2- 1.5*iqr2
    upperfence2 = q5+ 1.5*iqr2

    q6 = spad.groupby("Nrates")["Spad3"].quantile(0.25)
    q7 = spad.groupby("Nrates")["Spad3"].quantile(0.75)
    iqr3 = q3 - q1
    lowerfence3 = q6- 1.5*iqr3
    upperfence3 = q7+ 1.5*iqr3


    filtered = (group[(group["Spad1"] > lowerfence)  & (group["Spad1"] < upperfence)], 
                group[(group["Spad2"] > lowerfence2)  & (group["Spad2"] < upperfence2)], 
                group[(group["Spad3"] > lowerfence3)  & (group["Spad3"] < upperfence3)])

    return filtered
filtered = spad.groupby("Nrates").apply(filter_group).reset_index(drop=True)
```
